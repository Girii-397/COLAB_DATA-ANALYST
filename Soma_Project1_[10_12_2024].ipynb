```python
# ... (Your existing code)

# # **2. Handle missing values appropriately (drop or impute them).**

# Choose one of the following methods to handle missing values:

# Method 1: Drop rows with any missing values
# df = df.dropna()


# Method 2: Drop columns with any missing values
# df = df.dropna(axis=1)


# Method 3: Impute missing values with the mean of the column (for numerical columns)
for col in df.select_dtypes(include=['number']):  # Iterate only over numerical columns
    df[col] = df[col].fillna(df[col].mean())

# Method 4: Impute missing values with the median of the column (for numerical columns)
# for col in df.select_dtypes(include=['number']):
#     df[col] = df[col].fillna(df[col].median())


# Method 5: Impute missing values with the most frequent value (mode) (for categorical columns)
for col in df.select_dtypes(exclude=['number']):  # Iterate only over non-numerical columns
    df[col] = df[col].fillna(df[col].mode()[0]) # Use mode()[0] to get the first most frequent value

# Display the DataFrame after handling missing values
print("\nDataFrame after handling missing values:")
df
```

# **1 Create new columns:**

o Day of the week.

o Day of the month.

o Month.

o Year.

 o Season (Winter, Spring, Summer, Fall).

```python
# Ensure the date column is properly formatted (replace 'dteday' with your actual column name)
# Assuming the date column is named 'timestamp' based on global variables
df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')

# Extract day of the week
df['Day_of_Week'] = df['timestamp'].dt.day_name()

# Extract day of the month
df['Day_of_Month'] = df['timestamp'].dt.day

# Extract month
df['Month'] = df['timestamp'].dt.month_name()

# Extract year
df['Year'] = df['timestamp'].dt.year

# Define seasons based on months
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

df['Season'] = df['timestamp'].dt.month.apply(get_season)

# Display the updated DataFrame with new columns
df
```
